```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar el archivo CSV
archivo_csv = "Libro1.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Separar la fila de "Prob. Adopción"
df_prob_adopcion = df[df['Año'] == 'Prob. Adopción']

# Filtrar el resto de las filas para excluir "Prob. Adopción"
df_filtrado = df[df['Año'] != 'Prob. Adopción']

# Convertir los porcentajes en números flotantes y remover el símbolo '%'
df_filtrado = df_filtrado.replace('%', '', regex=True).replace('-', '0', regex=True).astype(float)

# Mostrar las primeras filas para verificar
df_filtrado.head()

```

```python


# Calcular la línea de tendencia
x = df_filtrado['Año']
y = df_filtrado['Python']
m, b = np.polyfit(x, y, 1) 
# Ajustar los datos a un polinomio de grado 2 (curva)
coeficientes = np.polyfit(x, y, 2)  # Grado 2
polinomio = np.poly1d(coeficientes)

# Generar valores de la línea de tendencia
x_linea = np.linspace(x.min(), x.max(), 100)  # Valores de X para la curva
y_linea = polinomio(x_linea)  # Valores de Y según el polinomio
# Visualizar la evolución de Python a lo largo de los años
plt.figure(figsize=(10, 6))
plt.scatter(x, y, marker='o', color='green', label="uso de Python (%)")
plt.plot(x_linea, y_linea, color='red', label="Línea de tendencia (curva)")  # Línea de tendencia curva

plt.title("Evolución de la uso de Python (2017-2023)", fontsize=16)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Porcentaje de uso (%)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

# el uso de python es cada vez mayor con el paso de los años
```
