```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Libro1.csv', delimiter=';')

# Convertir la columna 'Semana' a formato de fecha
data['Semana'] = pd.to_datetime(data['Semana'], format='%d/%m/%Y')

# Configurar el gráfico
plt.figure(figsize=(12, 8))

# Graficar cada columna de interés
plt.plot(data['Semana'], data['java: (Todo el mundo)'], label='Java', marker='o')
plt.plot(data['Semana'], data['C++: (Todo el mundo)'], label='C++', marker='o')
plt.plot(data['Semana'], data['python: (Todo el mundo)'], label='Python', marker='o')
plt.plot(data['Semana'], data['c#: (Todo el mundo)'], label='C#', marker='o')
plt.plot(data['Semana'], data['javaScript: (Todo el mundo)'], label='JavaScript', marker='o')

# Configurar los detalles del gráfico
plt.xlabel('Semana')
plt.ylabel('Millones de Busquedas')
plt.title('Busquedas Semanales por Lenguaje de Programación')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()



# Visualizar la evolución de python a lo largo de los años
plt.figure(figsize=(20, 6))
plt.scatter(data['Semana'], data['python: (Todo el mundo)'], marker='o', color='green')
plt.title("Busquedas de python (2019-2024)", fontsize=16)
plt.xlabel("Semana", fontsize=12)
plt.ylabel("Millones de Busquedas", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
#se presentan graficos para entender la data y al final de dará una hipotesis
```

```python
# Crear una nueva columna para el año
data['Año'] = data['Semana'].dt.year

# Promedio anual por lenguaje
data_anual = data.groupby('Año').mean()

# Graficar las tendencias anuales
plt.figure(figsize=(12, 8))
plt.plot(data_anual.index, data_anual['java: (Todo el mundo)'], label='Java', marker='o')
plt.plot(data_anual.index, data_anual['C++: (Todo el mundo)'], label='C++', marker='o')
plt.plot(data_anual.index, data_anual['python: (Todo el mundo)'], label='Python', marker='o')
plt.plot(data_anual.index, data_anual['c#: (Todo el mundo)'], label='C#', marker='o')
plt.plot(data_anual.index, data_anual['javaScript: (Todo el mundo)'], label='JavaScript', marker='o')

# Detalles del gráfico
plt.xlabel('Semanas')
plt.ylabel('Millones de Busquedas')
plt.title('Tendencias Anuales por BusquedasLenguaje de Programación')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Sumar las puntuaciones de todos los lenguajes
data['Suma Total'] = (data['java: (Todo el mundo)'] + 
                      data['C++: (Todo el mundo)'] + 
                      data['python: (Todo el mundo)'] + 
                      data['c#: (Todo el mundo)'] + 
                      data['javaScript: (Todo el mundo)'])

# Graficar la suma total de puntuaciones por semana
plt.figure(figsize=(12, 8))
plt.plot(data['Semana'], data['Suma Total'], label='Suma Total de Puntuaciones', marker='o', color='b')

# Detalles del gráfico
plt.xlabel('Semana')
plt.ylabel('Suma Total de busquedas')
plt.title('Suma Total de busquedas Semanales de Lenguajes de Programación')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crear una nueva columna para el año
data['Año'] = data['Semana'].dt.year

# Sumar las puntuaciones de todos los lenguajes por año
data_anual = data.groupby('Año').agg({
    'java: (Todo el mundo)': 'sum',
    'C++: (Todo el mundo)': 'sum',
    'python: (Todo el mundo)': 'sum',
    'c#: (Todo el mundo)': 'sum',
    'javaScript: (Todo el mundo)': 'sum'
})

# Crear una columna con la suma total por año
data_anual['Suma Total'] = data_anual.sum(axis=1)

# Graficar la suma total por año
plt.figure(figsize=(12, 8))
plt.plot(data_anual.index, data_anual['Suma Total'], label='Suma Total de busquedas por Año', marker='o', color='b')

# Detalles del gráfico
plt.xlabel('Año')
plt.ylabel('Suma Total de busquedas')
plt.title('Suma Total de busquedas Anuales de Lenguajes de Programación')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar el gráfico
plt.show()

#Hipotesis: en el 2022 hubo un aumento en las busquedas hechas en google relacionadas a varios lenguajes de programación
#Conclusión: Se realizo el analisis de estos datos sumando las busquedas de distintos lenguajes y nos dio la respuesta que a la hipotesis la cual es correcta
#en el 2022 hubo un aumento en las  busquedas hechas en google relacionadas a varios lenguajes de programación
```
