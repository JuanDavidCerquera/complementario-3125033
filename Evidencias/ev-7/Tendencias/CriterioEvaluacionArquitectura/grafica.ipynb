```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar el archivo CSV
archivo_csv = "dos.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Mostrar las primeras filas del archivo para verificar los datos
df.head()
```

```python

# Obtener la media por Arquitectura
df_mean = df.groupby('Arquitectura').mean()

# Categorías de la gráfica
categories = list(df_mean.columns)
N = len(categories)

# Ángulos para la gráfica de radar
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

# Iniciar la gráfica
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Dibujar una por una las líneas de cada Arquitectura
for index, row in df_mean.iterrows():
    values = row.tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=index)
    ax.fill(angles, values, alpha=0.1)

# Añadir etiquetas
plt.xticks(angles[:-1], categories)
ax.set_rlabel_position(0)
plt.yticks([1, 2, 5, 7, 10], ["1", "2", "5", "7", "10"], color="grey", size=7)
plt.ylim(0, 12)

# Añadir leyenda
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()


```
