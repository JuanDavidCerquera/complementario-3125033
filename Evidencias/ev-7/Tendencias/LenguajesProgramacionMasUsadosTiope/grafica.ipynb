```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

archivo_csv = "Libro5.csv" 
df = pd.read_csv(archivo_csv, sep=';')
df.head()
```

```python
# Crear el grfico de barras para las calificaciones de los lenguajes de programacin
plt.figure(figsize=(12, 8))  # Tamao del grfico

# Crear la grfica de barras usando los datos de calificaciones
plt.barh(df['Programming Language'], df['Ratings'], color='teal')

# Aadir etiquetas y ttulo
plt.xlabel('Calificacin (%)')
plt.ylabel('Lenguaje de Programacin')
plt.title('Calificacin de Lenguajes de Programacin en Sep-24')

# Mostrar la grfica
plt.show()
```

```python
# Asignar valores numricos a los lenguajes para calcular la lnea de tendencia
df['nombre_numerico'] = range(len(df))  # Crear ndices numricos para los lenguajes
y= df['nombre_numerico']
x = df['Ratings']
point = df['Change']  


# Ajustar los datos a un polinomio de grado 2 (curva)
coeficientes = np.polyfit(x, y, 2)  # Grado 2
polinomio = np.poly1d(coeficientes)

# Generar valores de la lnea de tendencia
x_linea = np.linspace(x.min(), x.max(), 100)  # Valores de X para la curva
y_linea = polinomio(x_linea)  # Valores de Y segn el polinomio


plt.figure(figsize=(10, 13))
# Agregar etiquetas con el nombre del trabajo en cada punto
for i in range(len(point)):
    plt.annotate(point[i], (x[i], y[i]), textcoords="offset points", xytext=(5, 5), ha='left', fontsize=9)
plt.scatter(x, y, marker='o', color='blue', label="porcentaje de cambio del uso del lenguaje")
plt.plot(x_linea, y_linea, color='red', label="Lnea de tendencia (curva)")  # Lnea de tendencia curva
# Configurar el eje X con las etiquetas originales
plt.yticks(ticks=df['nombre_numerico'], labels=df['Programming Language'], rotation=45, ha='right')


plt.title("Evolucin del uso de los lenguajes de programacin", fontsize=16)
plt.xlabel("Porcentaje de uso (%)", fontsize=12)
plt.ylabel("Lenguaje", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```
