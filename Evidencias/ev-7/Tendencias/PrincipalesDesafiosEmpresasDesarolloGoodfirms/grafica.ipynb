```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Intentar leer el archivo CSV con varias codificaciones
df = pd.read_csv('lib.csv', sep=';')
df.head()
```

```python
# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(df['Challenge'], df['Percentage'], color='skyblue')

# Configurar el gráfico
plt.xlabel('Desafio')
plt.ylabel('Porcentaje')
plt.title('Desafios Empresas')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python

df['nombre_numerico'] = range(len(df))  # Crear índices numéricos para los lenguajes
y= df['nombre_numerico']
x = df['Percentage']


# Ajustar los datos a un polinomio de grado 2 (curva)
coeficientes = np.polyfit(x, y, 2)  # Grado 2
polinomio = np.poly1d(coeficientes)

# Generar valores de la línea de tendencia
x_linea = np.linspace(x.min(), x.max(), 100)  # Valores de X para la curva
y_linea = polinomio(x_linea)  # Valores de Y según el polinomio


plt.figure(figsize=(10, 13))
plt.scatter(x, y, marker='o', color='blue', label="Dificultad de categoria")
plt.plot(x_linea, y_linea, color='red', label="Línea de tendencia (curva)")  # Línea de tendencia curva
# Configurar el eje X con las etiquetas originales
plt.yticks(ticks=df['nombre_numerico'], labels=df['Challenge'], rotation=45, ha='right')


plt.title("Porcentaje de dificultades por Challenge", fontsize=16)
plt.xlabel("Porcentaje de dificultades encontradas (%)", fontsize=12)
plt.ylabel("Lenguaje", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```
