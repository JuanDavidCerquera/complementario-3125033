```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar el archivo CSV
archivo_csv = "Libro1.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Convertir los porcentajes a n√∫meros flotantes
df['Porcentaje'] = df['Porcentaje'].str.replace('%', '').astype(float)

# Mostrar las primeras filas del archivo para verificar los datos
df.head()
```

```python
# Crear una gr√°fica de pastel
plt.figure(figsize=(8, 8))  # Ajustar el tama√±o de la figura

plt.pie(df['Porcentaje'], labels=df['Lenguaje'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# A√±adir un t√≠tulo
plt.title('Porcentaje de uso de lenguajes de programaci√≥n')

# Mostrar la gr√°fica
plt.show()
```

```python
# Crear la gr√°fica de barras
plt.figure(figsize=(12, 8))  # Tama√±o del gr√°fico

# Ordenar los datos por porcentaje para mejor visualizaci√≥n
df_sorted = df.sort_values(by='Porcentaje', ascending=False)

# Crear el gr√°fico de barras
plt.barh(df_sorted['Lenguaje'], df_sorted['Porcentaje'], color='skyblue')

# A√±adir etiquetas y t√≠tulo
plt.xlabel('Porcentaje')
plt.ylabel('Lenguaje')
plt.title('Porcentaje de Uso de Tecnolog√≠as')

# Mostrar la gr√°fica
plt.show()
```

```python
# Asignar valores num√©ricos a los lenguajes para calcular la l√≠nea de tendencia
df['Lenguaje_numerico'] = range(len(df))  # Crear √≠ndices num√©ricos para los lenguajes
x = df['Lenguaje_numerico']
y = df['Porcentaje']

# Ajustar los datos a un polinomio de grado 2 (curva)
coeficientes = np.polyfit(x, y, 2)  # Grado 2
polinomio = np.poly1d(coeficientes)
# Configurar el eje X con las etiquetas originales
plt.xticks(ticks=df['Lenguaje_numerico'], labels=df['Lenguaje'], rotation=45, ha='right')

# Generar valores de la l√≠nea de tendencia
x_linea = np.linspace(x.min(), x.max(), 100)  # Valores de X para la curva
y_linea = polinomio(x_linea)  # Valores de Y seg√∫n el polinomio
plt.plot(x_linea, y_linea, color='red', label="L√≠nea de tendencia (curva)")  # L√≠nea de tendencia curva
plt.scatter(x, y, marker='o', color='blue', label="porcentaje del uso del lenguaje")


plt.title("Evoluci√≥n del uso de los lenguajes de programaci√≥n", fontsize=16)
plt.xlabel("Lenguaje", fontsize=12)
plt.ylabel("Porcentaje de uso (%)", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

#Hipotesis: Se puede evidenciar gran diferencia entre el uso de python y de los demas lenguajes de programaci√≥n
#Conclusion: En la grafica se puede evidenciar la diferencia del uso de cada lenguaje y se puede apreciar que python esta muy por encima de los demas lenguajes

```
