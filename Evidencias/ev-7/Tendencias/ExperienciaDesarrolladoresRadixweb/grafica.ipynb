```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar el archivo CSV
archivo_csv = "Libro6.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Mostrar las primeras filas del archivo para verificar los datos
df.head()
```

```python
# Asegurarse de que las columnas tengan los nombres correctos
x = df['Median Base Salary']
y = df['Job Satisfaction']
job_titles = df['Job Title']  # Columna con los nombres de los trabajos

# Ajustar un polinomio de grado 2 (curva)
coeficientes = np.polyfit(x, y, 2)  # Grado 2
polinomio = np.poly1d(coeficientes)

# Generar valores de la curva
x_linea = np.linspace(x.min(), x.max(), 100)
y_linea = polinomio(x_linea)

# Visualizar la curva de tendencia
plt.figure(figsize=(12, 8))
plt.scatter(x, y, marker='o', color='green', label="Puestos de trabajo")

# Agregar etiquetas con el nombre del trabajo en cada punto
for i in range(len(job_titles)):
    plt.annotate(job_titles[i], (x[i], y[i]), textcoords="offset points", xytext=(5, 5), ha='left', fontsize=9)

# Graficar la lnea de tendencia (curva)
plt.plot(x_linea, y_linea, color='red', label="Lnea de tendencia (curva)")

# Configuracin del grfico
plt.title("Clasificando buenos trabajos", fontsize=16)
plt.xlabel("Salario Base Mediano", fontsize=12)
plt.ylabel("Nivel de Satisfaccin", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

#hipotesis : El salario afecta la satisfaccion que tienen las personas en su puesto de trabajo
#conclucin: Se entiende que al contrario de la hipotesis propuesta el salario no afecta la satisfaccion que tienen las personas en su puesto de trabajo
```
