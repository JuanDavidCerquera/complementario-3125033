```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar el archivo CSV
archivo_csv = "Libro7.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Mostrar las primeras filas del archivo para verificar los datos
df.head()
```

```python
df.plot(kind='bar', x='Category', y=['2023 Spending', '2024 Spending'], color=['blue', 'green'])

# Aadir etiquetas y ttulo
plt.xlabel('Categora')
plt.ylabel('Gasto (USD)')
plt.title('Comparacin de gastos IT en 2023 vs 2024')

# Mostrar la grfica
plt.show()



# Crear una grfica de barras para comparar el crecimiento porcentual de 2023 y 2024
df.plot(kind='bar', x='Category', y=['2023 Growth (%)', '2024 Growth (%)'], color=['red', 'orange'])

# Aadir etiquetas y ttulo
plt.xlabel('Categora')
plt.ylabel('Crecimiento (%)')
plt.title('Comparacin de crecimiento IT en 2023 vs 2024')

# Mostrar la grfica
plt.show()
```

```python
# Asegurarse de que las columnas tengan los nombres correctos
x = df['2023 Growth (%)']
y = df['2024 Growth (%)']
job_titles = df['Category']  # Columna con los nombres de los trabajos

# Ajustar un polinomio de grado 2 (curva)
coeficientes = np.polyfit(x, y, 2)  # Grado 2
polinomio = np.poly1d(coeficientes)

# Generar valores de la curva
x_linea = np.linspace(x.min(), x.max(), 100)
y_linea = polinomio(x_linea)

# Visualizar la curva de tendencia
plt.figure(figsize=(12, 8))
plt.scatter(x, y, marker='o', color='blue', label="Categoria IT")

# Agregar etiquetas con el nombre del trabajo en cada punto
for i in range(len(job_titles)):
    plt.annotate(job_titles[i], (x[i], y[i]), textcoords="offset points", xytext=(5, 5), ha='left', fontsize=9)

# Graficar la lnea de tendencia (curva)
plt.plot(x_linea, y_linea, color='red', label="Lnea de tendencia (curva)")

# Configuracin del grfico
plt.title("Crecimiento 2023 vs Crecimiento 2024", fontsize=16)
plt.xlabel("2023 Growth (%)", fontsize=12)
plt.ylabel("2024 Growth (%)", fontsize=12)
plt.legend()
plt.show()
# Hipotesis: El crecimiento que tiene cada categoria de IT e similar es ambos aos
# Concluson: Se entiende que no todas las categorias siguen la tendencia
```
