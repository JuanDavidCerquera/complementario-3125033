```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar el archivo CSV
archivo_csv = "Libro1.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')


# Mostrar las primeras filas del archivo para verificar
df.head()
```

```python
# Crear una gráfica de pastel
plt.figure(figsize=(8, 8))  # Ajustar el tamaño de la figura

plt.pie(df['Percentage'], labels=df['Technology'], colors=plt.cm.Paired.colors)

# Añadir un título
plt.title('Porcentaje de uso de tecnologias mas usadas')

# Mostrar la gráfica
plt.show()
```

```python
# Crear la gráfica de barras
plt.figure(figsize=(12, 8))  # Tamaño del gráfico

# Ordenar los datos por porcentaje para mejor visualización
df_sorted = df.sort_values(by='Percentage', ascending=False)

# Crear el gráfico de barras
plt.barh(df_sorted['Technology'], df_sorted['Percentage'], color='skyblue')

# Añadir etiquetas y título
plt.xlabel('Porcentaje')
plt.ylabel('Tecnología')
plt.title('Porcentaje de Uso de Tecnologías')

# Mostrar la gráfica
plt.show()
```

```python
# Asignar valores numéricos a los lenguajes para calcular la línea de tendencia
df['Technology_numerico'] = range(len(df))  # Crear índices numéricos para los lenguajes
y= df['Technology_numerico']
x = df['Percentage']


# Ajustar los datos a un polinomio de grado 2 (curva)
coeficientes = np.polyfit(x, y, 2)  # Grado 2
polinomio = np.poly1d(coeficientes)

# Generar valores de la línea de tendencia
x_linea = np.linspace(x.min(), x.max(), 100)  # Valores de X para la curva
y_linea = polinomio(x_linea)  # Valores de Y según el polinomio


plt.figure(figsize=(10, 13))
plt.scatter(x, y, marker='o', color='blue', label="porcentaje del uso del lenguaje")
plt.plot(x_linea, y_linea, color='red', label="Línea de tendencia (curva)")  # Línea de tendencia curva
# Configurar el eje X con las etiquetas originales
plt.yticks(ticks=df['Technology_numerico'], labels=df['Technology'], rotation=45, ha='right')


plt.title("Evolución del uso de los lenguajes de programación", fontsize=16)
plt.xlabel("Lenguaje", fontsize=12)
plt.ylabel("Porcentaje de uso (%)", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```
